```python
import numpy as np
import matplotlib.pyplot as pit
```

```python
# Data Structure : [restaurant_id, 2021, 2022, 2023, 2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # Paradise Biryani
    [2, 120000, 140000, 160000, 190000], # Bejing Bites
    [3, 200000, 230000, 260000, 300000], # Pizza Hub
    [4, 180000, 210000, 240000, 227000], # Burger Pont
    [5, 160000, 185000, 205000, 230000]  # Chai Point
])

print("\n How Many (Row, Column) : ", sales_data.shape)
print("\nSample Data For 1st Three restaurant : ", sales_data[0:3])

# sale_data(row_start_index : row_end_index, Column_start_index : column_end_index)
print("\nSample Data For 1st Three restaurant with starting two year with restaurant Id : ", sales_data[0:3, 0:3])
```

```python
# Total Sales Per Year

# print("Total Sale of 2021 to 2024 of every restaurant : ", np.sum(sales_data, axis=0)[1:])
print("Total Sale of 2021 to 2024 every restaurant : ", np.sum(sales_data[:,1:], axis=0))

print("Total Sale From 2021 to 2024 of perticuler restaurant : ", np.sum(sales_data[:,1:], axis=1))

print("Minimum Sale of retuarent : ", np.min(sales_data[:, 1:], axis=1))

print("Maxmimum Sale of year : ", np.max(sales_data[:, 1:], axis=0) )

print("Average Sales of restaurant : ", np.mean(sales_data[:, 1:], axis=1))

# cumulative sale mean = first_year_Sale, first_year_Sale+second_year, first+second+third ans so on..
cumulative_Sum = np.cumsum(sales_data[:,1:], axis=1)
print("cumulative sum of restaurant : ", cumulative_Sum)
```

```python
# Some Section Of Matplotlib

pit.figure(figsize=(10,5))
pit.plot(np.mean(cumulative_Sum, axis=0))
pit.title("Average Cumulative Sale Across all Resturant")
pit.xlabel("Year")
pit.ylabel("Sales")
pit.grid(True)
pit.show
```

# Vectors Some Topic

```python
vector_1 = np.array([1,2,3,4,5])
vector_2 = np.array([7,8,9,6,1])

vector_sum = vector_1 + vector_2
print(vector_sum)

vector_multiplication = vector_1 * vector_2
print(vector_multiplication)

vector_devide = vector_1/vector_2
print(vector_devide)

dot_product = np.dot(vector_1, vector_2)
print(dot_product)

angle = np.arccos(np.dot(vector_1, vector_2) / (np.linalg.norm(vector_1) * np.linalg.norm(vector_2)))
print(angle)
```
