```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array_1 = np.array([[1,2],[3,4]])
array_2 = np.random.random((3,3))
array_3 = np.zeros((3,4))
```

```python
# save Any Array in a numpy flat file
# Basically we save .npy file in this file we save array/marix of any picture, video and etc.
np.save("Array_1.npy", array_1)
np.save("Array_2.npy",array_2)
np.save("Array_3.npy",array_3)

print(np.load('Array_1.npy'))
```

```python
# numpy-logo.npy in this file logo of numpy save in matrix format
# and we can load this matrix and perform some operation
# Many AI tools like ChatGPT, when we give any picture to the ChatGPT and give some promt then ChatGPT save This file in pixle or matrix and given promt it work on matrix and give result

try:
    logo = np.load("numpy-logo.npy")
    # print(logo)

    # Display

    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title('NumPy Logo')
    plt.grid(False)

    # for dark logo
    dark_logo = 1- logo
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy Dark Logo")
    plt.grid(False)
except FileNotFoundError:
    print("File Not Found")
```
